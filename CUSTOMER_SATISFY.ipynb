```python
import pandas as pd

# Step 1: Define available product categories
product_categories = ["Electronics", "Clothing", "Home Appliances", "Books", "Furniture"]

# Step 2: Ask user to select a product category
print("Available Product Categories:")
for i, category in enumerate(product_categories, 1):
    print(f"{i}. {category}")

selected_index = int(input("\nSelect a product category by entering the corresponding number: ")) - 1
selected_category = product_categories[selected_index]
print(f"\nYou selected: {selected_category}")

# Step 3: Collect customer ratings for the selected product
ratings = []
num_customers = int(input(f"\nHow many customers are providing ratings for {selected_category}? "))

for i in range(1, num_customers + 1):
    rating = float(input(f"Enter rating from Customer {i} (1-5): "))
    ratings.append(rating)

# Step 4: Store data in a DataFrame
data = {"Product_Category": [selected_category] * num_customers, "Satisfaction_Score": ratings}
df = pd.DataFrame(data)

# Step 5: Calculate the average satisfaction score for the selected product
average_score = df["Satisfaction_Score"].mean()
print(f"\nAverage Satisfaction Score for {selected_category}: {average_score:.2f}")

# Step 6: Optional - Visualize the ratings distribution
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.histplot(df["Satisfaction_Score"], bins=5, kde=True)
plt.title(f"Satisfaction Scores for {selected_category}")
plt.xlabel("Satisfaction Score")
plt.ylabel("Frequency")
plt.show()
```
